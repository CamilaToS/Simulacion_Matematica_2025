```python
import numpy as np
```

```python
def actividad(val):
  c=range(2,val)
  lista=[]
  for i in c:
    d=np.arange(2,i)
    m=(i%d)==0
    if not m.any():
      lista.append(i)
  return lista


actividad(100) #:)
```

```python
def primos_menores_que(n):
    if n < 2:
        return []

    # Inicializa una lista de booleanos: True significa "posiblemente primo"
    es_primo = [True] * n
    es_primo[0:2] = [False, False]  # 0 y 1 no son primos

    for i in range(2, int(n**0.5) + 1):
        if es_primo[i]:
            # Marca todos los múltiplos de i como no primos
            for j in range(i*i, n, i):
                es_primo[j] = False

    # Extrae los índices que siguen siendo True (es decir, los primos)
    return [i for i, primo in enumerate(es_primo) if primo]

# Ejemplo de uso
print(primos_menores_que(100))
```

```python
from sympy import primerange

def primos_sympy(n):
    return list(primerange(2, n))

# Ejemplo de uso
print(primos_sympy(100))
```
