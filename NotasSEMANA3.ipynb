### PyPlot de matplotlib

<img style="float: right; margin: 0px 0px 15px 15px;" src="https://matplotlib.org/_static/logo2.svg" width="300px" height="100px" />

El módulo `PyPlot` de la librería `matplotlib` contiene funciones que nos permite generar una gran cantidad de gráficas rápidamente. Las funciones de este módulo están escritas con el mismo nombre que las funciones para graficar en `Matlab`.

**Referencia:**
- https://matplotlib.org/api/pyplot_summary.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,2*np.pi,500) #Genera un arreglo de 500 elementos entre 0 y 2pi
y=np.sin(x) #Evalúa la función seno en cada uno de los 500 elementos de x
```

```python
plt.plot(x,y) #Gráfica x vs y
plt.grid() #Cuadrícula al fondo
```

```python
plt.plot(x,y,'r') #Color en el tercer elemento
plt.xlabel('x') #Nombra el eje x
plt.ylabel('y')
plt.title('sin(x)') #Título de la gráfica
plt.grid()
```

```python
#Para buscar más opciones de colores, figuras, etc, puedo encontrarlas en la opción ayuda:
#plt.plot?
```

```python
plt.plot(x,y,':g') #graficar seno como linea punteada verde
puntos=np.array([0,np.pi,2*np.pi]) #puntos de interés
plt.plot(puntos,np.sin(puntos),'xr') #marca estos puntos con una x roja
plt.grid()
```

```python
plt.plot(x,y,':g',lw=4) #lw= linewidth
puntos=np.array([0,np.pi,2*np.pi])
plt.plot(puntos,np.sin(puntos),'xr',ms=20) #'ms'='markersize'
plt.grid()
```

```python
plt.plot(x,np.sin(x),
         x,np.cos(x),
         x,np.exp(-0.5*x)*np.sin(2*x))
plt.legend(['seno','coseno','seno atenuado'],bbox_to_anchor=(1.01,0.5),loc=6) #en este caso, loc es el punto en el que se posicionan las coordenadas
#en y pongo 0.5 y significa que estará en el centro de la y, ya que 0 es arriba y 1 es abajo.
#En x un valor mayor a 1 me especifica que lo saque de la imagen, siendo 0 a la izquierda y 1 a la derecha de la figura
plt.grid()
```

```python
plt.plot(x,np.sin(x),x,np.cos(x),x,np.tan(x))
#plt.ylim(-5,5)     #Límites en eje y
#plt.xlim(0,10)     #límites en x
plt.axis([0,2*np.pi,-5,5]) #Límite para ambos ejes: (x_min, x_max, y_min, y_max)
plt.grid()
```
