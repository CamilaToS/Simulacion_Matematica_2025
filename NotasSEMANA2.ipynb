```python
def actividad(val):
  c=range(2,val)
  lista=[]
  for i in c:
    d=np.arange(2,i)
    m=(i%d)==0
    if not m.any():
      lista.append(i)
  return lista


actividad(100) #WOOW
```

```python
#definir una variable simbólica:
x=sym.var('x') #variable dependiente de x
f=sym.sin(x)*sym.exp(-x)
f
```

```python
sym.diff(f,x) #Derivar f con respecto de x
```

```python
sym.diff(f) #lo mismo, pues sólo tengo una variable en f
```

```python
#Derivar más de una vez:
sym.diff(f,x,2)  #Derivando dos veces
```

```python
#Para encontrar una raíz: cuando f(x)=0
sym.solve(f)
```

```python
#Para graficar:
sym.plot(f,(x,-sym.pi,sym.pi)) #grafico f, limitando x de -pi a pi
```

```python
#Limitada a 2pi positivo
sym.plot(f,(x,0,2*sym.pi)) #Rango limitado
```
